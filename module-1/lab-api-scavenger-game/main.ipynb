```python
resp5 = req.get('https://api.github.com/repos/ironhack-datalabs/scavenger/contents')
results5 = resp5.json()
results5
```

```python
mensaje = []
for i in range(1,len(results5)):
    mensaje.append(results5[i]['url'])
mensaje
```

```python
mensaje_carpeta = []
for carp in mensaje:
    respcarp6 = req.get(carp)
    resultscarp6 = respcarp6.json()
    for j in resultscarp6:
        if 'scavengerhunt' in j['name']:
            mensaje_carpeta.append([j['name'],j['download_url']])

mensaje_carpeta
```

```python
mensaje_carpeta = sorted(mensaje_carpeta)
mensaje_carpeta
```

```python
mensajeoculto = ""
for m in mensaje_carpeta:
    respfinal = req.get(m[1])
    mensajeoculto=mensajeoculto+(respfinal.text.replace('\n',' '))
print(f'Hidden message: {mensajeoculto}')
```
