```python
import random
import time
import sys
tir_ale=["M. Ballack","L. Podolski","T. Muller","T. Kroos","P. Lahm","M. Klose","T. Werner","M. Goretzka","F. Beckenbauer","J. Klinsmann","K. Havertz"]
tir_mex=["C. Blanco","R. Márquez","P. Pardo","C. Vela","R. Jiménez","J. Borguetti","A. Guardado","H. Sánchez","L. Hernández","Zague","J. Campos"]
tir_arg=["L. Messi","D. Maradona","C. Tévez","C. Lòpez","S. Aguero","P. Dybala","J. Batistuta","M. Kempes","J. Riquelme","A. Dí María","H. Crespo"]
tir_bra=["Ronaldo","Neymar JR","Ronaldinho","Kaka","Roberto Carlos","Pele","Lucio","Cafú","Dani Alves","Socrates","Garrincha"]
tir_fra=["Z. Zidane","T. Henry","K. Benzema","M. Platini","E. Cantona","F. Ribéry","K. Mbappé","P. Pogba","N. Makelele","M. Papin","A. Griezmann"]
print("Bienvenido(a), ¿Estas listo(a) para jugar una tanda de penalties?")
print("     ________________________________     ")
print("     |                              |     ")
print("     |                              |     ")
print("     |      Tanda    Penalties      |     ")
print("     |                              |     ")
print("_____|______________________________|_____")
print("")
print("")

print("¿Que equipo te gustaría usar?")
paises=["MEX","ARG","BRA","FRA","ALE"]
des_cpu=random.choice(paises)

des_jugador=input("Elige entre: MEX ARG BRA FRA ALE ")
while des_jugador not in paises:
    print("Elige un equipo válido o que no este tomado por tu contrincante ") ##Si es que se escribe algun equipo que no esta en la lista, o lo tiene el cpu

    des_jugador=input("Aquí ")
while des_jugador==des_cpu:
    print("El equipo esta tomado por tu contrincante")
    des_jugador=input("Elige otro equipo ")
print("")
print("La tanda será ",des_jugador,"vs ",des_cpu)


movs=["I","C","D","I","C","D","I","C","D","I","C","D""P","P"] ##probabilidad de cada movimiento, de 2 de cada 14 tiros pueden ser poste
    
total_penales=0
jugador_marc=0
cpu_marc=0
print("")
    
def penal_a_favor(): ##definiendo la funcion cuando el usuario tira un penal
    global total_penales
    global jugador_marc    
    total_penales+=1
    direccion_penal=input("Elige la dirección del tiro I,C,D:")
    while direccion_penal not in movs:
        print("Elige un tiro válido")
        direccion_penal=input("Aquí ")
    parada_cpu=random.choice(movs)
    if parada_cpu=="P":
        print("Mala suerte, le diste al poste!")
    else:
        if ((direccion_penal=="I" and parada_cpu=="D") | (direccion_penal=="D" and parada_cpu=="I") | (direccion_penal=="C" and parada_cpu=="C")):
            for remaining in range(2, 0, -1):  ##cuenta regresiva para dar emocion jajajaj
                sys.stdout.write("\r")
                sys.stdout.write("se prepara {:2d}".format(remaining))
                sys.stdout.flush()
                time.sleep(1)
            sys.stdout.write("\rMala suerte, el portero detuvo tu tiro :(           \n")
            
        else:
            for remaining in range(2, 0, -1):
                sys.stdout.write("\r")
                sys.stdout.write("se prepara {:2d}".format(remaining))
                sys.stdout.flush()
                time.sleep(1)

            sys.stdout.write("\rGOOOOOOOOOOOOL!!!! de: \n")
            if des_jugador=="MEX":
                print(random.choice(tir_mex))
            elif des_jugador=="ARG":
                print(random.choice(tir_arg))
            elif des_jugador=="BRA":
                print(random.choice(tir_bra))
            elif des_jugador=="FRA":
                print(random.choice(tir_fra))
            else:
                print(random.choice(tir_ale))
            jugador_marc+=1  ##suma al marcador
                
        
print("")

def penal_en_contra():  ##definiendo la funcion cuando el usuario es portero
    global total_penales
    global cpu_marc
    total_penales+=1
    dir_portero=input("Elige a donde te quieres lanzar I,C,D:")
    while dir_portero not in movs:
        print("Elige un lado válido")
        direccion_penal=input("Aquí ")
    tiro_cpu=random.choice(movs)
    if tiro_cpu=="P":
        print("Te salvaste! El tiro le dio al palo!!!")
    else:
        if ((tiro_cpu=="I" and dir_portero=="D") | (tiro_cpu=="D" and dir_portero=="I") | (tiro_cpu=="C" and dir_portero=="C")):
            for remaining in range(2, 0, -1):
                sys.stdout.write("\r")
                sys.stdout.write("se prepara {:2d}".format(remaining))
                sys.stdout.flush()
                time.sleep(1)
            sys.stdout.write("\rBien hecho!!! Detuviste el tiro!!!           \n")
            
        else:
                
            for remaining in range(2, 0, -1):
                sys.stdout.write("\r")
                sys.stdout.write("se prepara {:2d}".format(remaining))
                sys.stdout.flush()
                time.sleep(1)

            sys.stdout.write("\rMala suerte. GOOOOL de:            \n")
            if des_cpu=="MEX":
                print(random.choice(tir_mex))
            elif des_cpu=="ARG":
                print(random.choice(tir_arg))
            elif des_cpu=="BRA":
                print(random.choice(tir_bra))
            elif des_cpu=="FRA":
                print(random.choice(tir_fra))
            else:
                print(random.choice(tir_ale))
            cpu_marc+=1
print("") 

def marcador():  ##marcador parcial 
    print("EL MARCADOR ES:",des_jugador ,(jugador_marc),"vs",str(cpu_marc), des_cpu)

print("")
def Marcador_final():
    print("MARCADOR FINAL ",des_jugador,"",jugador_marc,"vs",cpu_marc,"",des_cpu) 
    if jugador_marc>cpu_marc:
        print ("Felicidades ganaste con ",des_jugador)
    elif jugador_marc<cpu_marc:
        print("Perdiste contra ",des_cpu,":(")
    else:
        print("EMPATE!!!")
            
         
            
    
while total_penales<10: ##5 penales por lado
    penal_a_favor()
    penal_en_contra()
    print("")
    marcador()
print("")
Marcador_final()   
   

    
    


    
```
