```python
#Im giving a list of movies
movies = "titanic avatar godfather starwars psycho jaws kingkong casablanca rocky et matrix exorcist ghostbusters scarface braveheart beetlejuice goodfellas rambo bambi footlose poltergeist gladiator grease seven tron ghost metropolis hangover borat fargo".split() 
```

```python
#Im telling the program  how to choose from my list
def get_randomly_movies(movies):
    movie_index = random.randint(0, len(movies) - 1)
    return movies[movie_index]
```

```python
#I did this step by step, but put it all together at the end, so it wont tell me that I have not defined some of the varients.


#Please show me my hanged person this way
def show_image (hanged_images, wrong_letter, right_letter, secret_word):
    print (hanged_images[len(wrong_letter)])
    print()
    
    print("Wrong letter", end = "")
    for letter in wrong_letter:
        print(letter, end = "")
    print ()
    
    empty_space = " "*len(secret_word)

    for i in range (len(secret_word)):
        if secret_word[i] in right_letter:
            empty_space = empty_space[:i] + secret_word[i] + empty_space[i+1:]
        

    for letter in empty_space:
        print (letter, end ="")
    print ()

#The way to get it is with inputs
def get_try (tried_letter):
    while True:
        print ("Guess a word of an iconic movie")
        tried = input()
        tried = tried.lower()
        if len (tried) != 1:
            print ("Please introduce a word")
        elif tried in tried_letter:
            print("You have tried that one before, try another")
        elif tried not in "abcdefghijklmn√±opqrstuvwxyz":
            print("Please try a LETTER")
        else:
            return tried
        
#You may be wrong so, Im planning for you to play again        
def play_again():
    print("You want to play again?")
    return input ().startswith ("y")


print ("HANGED")
wrong_letter = ""
right_letter = ""         
secret_word = get_randomly_movies(movies)
end_of_game = False

# Now lets define what is wrong and right
while True:
    show_image(hanged_images,wrong_letter, right_letter,secret_word)
    
    tried = get_try(wrong_letter + right_letter)
    if tried in secret_word:
        right_letter = right_letter + tried
        
        found_all_letters = True
        for i in range(len(secret_word)):
            if secret_word[i] not in right_letter:
                found_all_letters= False
                break
        if found_all_letters:
            print ("Crack!, The secret movies is:" + secret_word + "You won")
            end_of_game = True
    else:
        wrong_letter = wrong_letter + tried
        
        if len(wrong_letter) == len (hanged_images)-1:
            show_image (hanged_images, wrong_letter, right_letter, secret_word)
            print ("Ooopsie")
            end_of_game = True
            
    if end_of_game:
        if play_again ():
            wrong_letter = ""
            right_letter = ""         
            secret_word = get_randomly_movies(movies)
            end_of_game = False
        else:
            break 
            
        
```
