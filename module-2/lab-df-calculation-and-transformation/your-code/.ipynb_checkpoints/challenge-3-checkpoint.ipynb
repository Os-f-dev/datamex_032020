# Q1: How to identify VIP & Preferred Customers?

We start by importing all the required libraries:

Next, import `Orders.csv` from the "subsetting" lab folder into a dataframe variable called `orders`. Print the head of `orders` to overview the data:

"Identify VIP and Preferred Customers" is the non-technical goal of your boss. You need to translate that goal into technical languages that data analysts use:

## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
#step 1: get a DF of total spent per customer
customers_amount_spent = orders.groupby("CustomerID").agg({"amount_spent":"sum"})
customers_amount_spent
```

```python
#step 2: define percentiles
preferred = customers_amount_spent["amount_spent"].quantile(0.75)
vip = customers_amount_spent["amount_spent"].quantile(0.95)
```

```python
#step 3: label customers
def label(x):
    if x>=vip:
        return "VIP"
    elif x>=preferred:
        return "Preferred" 
    else:
        return "Normal"
    
customers_amount_spent["CustomerLabel"] = customers_amount_spent["amount_spent"].apply(lambda x: label(x))
```

```python
#step 4: we create a copy of customers_amount_spent to merge with orders df.

cust_label = customers_amount_spent[["CustomerLabel"]].copy()
```

```python
#step 5: merge dataframes

orders = pd.merge(orders, cust_label, on="CustomerID")
```

# Q2: How to identify which country has the most VIP Customers?

# Q3: How to identify which country has the most VIP+Preferred Customers combined?

Provide your solution for Q2 below:

```python
# Q2: How to identify which country has the most VIP Customers?
vip_country = orders[["CustomerID","Country","CustomerLabel"]].groupby(["CustomerLabel","Country"]).agg({"count"}).loc["VIP"].reset_index()

vip_country[vip_country[('CustomerID', 'count')]==vip_country[('CustomerID', 'count')].max()]


```

```python
#Q3: How to identify which country has the most VIP+Preferred Customers combined?
vip_country = orders[["CustomerID","Country","CustomerLabel"]].groupby(["CustomerLabel","Country"]).agg({"count"}).loc["VIP"].reset_index()
vip_country.columns = vip_country.columns.droplevel(1)
vip_country.columns = ["Country","VIP Customers"]

preferred_country = orders[["CustomerID","Country","CustomerLabel"]].groupby(["CustomerLabel","Country"]).agg({"count"}).loc["Preferred"].reset_index()
preferred_country.columns = preferred_country.columns.droplevel(1)
preferred_country.columns =["Country","Preferred Customers"]

vip_pref_cust = pd.merge(vip_country,preferred_country, on="Country")
vip_pref_cust["Sum"] = vip_pref_cust["VIP Customers"]+vip_pref_cust["Preferred Customers"]
vip_pref_cust[vip_pref_cust["Sum"]==vip_pref_cust["Sum"].max()]
```
