#### Using Seaborn plot the correlation matrix of various features.

*Hint: search how to use `heatmap`*

```python
#your code here
corr = titanic.corr()
mask = np.triu(np.ones_like(corr, dtype=np.bool))
cmap = sns.diverging_palette(220,20,as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, center=0, square=True, cbar_kws={'shrink':0.5},annot=True)
plt.show();
```

#### According to the previous plot, what are the most correlated feature? Comment.

SibSp & Parch

#### Use the appropriate plot to display the distributions of  `Age` in relation to  `Gender`.

*Hint: use Facetgrid to display the distribution of `Age` for each categorical value of `Gender`*

```python
#your code here
sns.kdeplot(titanic['Age'][titanic['Gender']=='female'], color='lightcoral',shade=True)
sns.kdeplot(titanic['Age'][titanic['Gender']=='male'], color='teal',shade=True)
plt.legend(['Female', 'Male'])
plt.show();
```
