# Plotting Multiple Data Series

Complete the following set of exercises to solidify your knowledge of plotting multiple data series with pandas, matplotlib, and seaborn. Part of the challenge that comes with plotting multiple data series is transforming the data into the form needed to visualize it like you want. For some of the exercises in this lab, you will need to transform the data into the form most appropriate for generating the visualization and then create the plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import numpy as np

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_csv('../data/liquor_store_sales.csv')
data.head()
```

## 1. Create a bar chart with bars for total Retail Sales, Retail Transfers, and Warehouse Sales by Item Type.

```python
item_types = data.groupby("ItemType")[["RetailSales","RetailTransfers","WarehouseSales"]].sum().reset_index()
item_types.plot(kind='bar',x="ItemType",y=["RetailSales","RetailTransfers","WarehouseSales"])
plt.show();
```

## 2. Create a horizontal bar chart showing sales mix for the top 10 suppliers with the most total sales. 

```python
suppliers = data[["Supplier","RetailSales","WarehouseSales"]]
suppliers["Total sales"] = suppliers["RetailSales"]+suppliers["WarehouseSales"]
top_suppliers = suppliers.groupby("Supplier")[["Total sales"]].sum().sort_values("Total sales",ascending=False).reset_index().head(10)
top_suppliers.plot.barh("Supplier","Total sales")
plt.show();
```

## 3. Create a multi-line chart that shows average Retail Sales, Retail Transfers, and Warehouse Sales per month over time.

```python
def month_year(x):
    month = str(int(x["Month"]))
    year = str(int(x["Year"]))
    return f"{month}-{year}"

month_sales = data[["Year","Month","RetailSales","RetailTransfers","WarehouseSales"]].groupby(["Year","Month"]).mean().reset_index()
month_sales["Month-Year"] = month_sales.apply(lambda x: month_year(x), axis=1)
```

```python
plt.figure(figsize=(14, 5))
plt.plot("Month-Year","RetailSales", data=month_sales)
plt.plot("Month-Year","RetailTransfers", data=month_sales)
plt.plot("Month-Year","WarehouseSales", data=month_sales)
plt.legend();
```

## 4. Plot the same information as above but as a bar chart.

```python
month_sales.plot(kind='bar',x="Month-Year",y=["RetailSales","RetailTransfers","WarehouseSales"])
plt.show();
```

## 5. Create a multi-line chart that shows Retail Sales summed by Item Type over time (Year & Month).

*Hint: There should be a line representing each Item Type.*

```python
item_sales = data.pivot_table(index = ["Year","Month"],
                           columns = ["ItemType"],
                           values = ["RetailSales"],
                           aggfunc = "sum")

item_sales.plot();
```

## 6. Plot the same information as above but as a bar chart.

```python
item_sales.plot(kind='bar');
```

## 7. Create a scatter plot showing the relationship between Retail Sales (x-axis) and Retail Transfers (y-axis) with the plot points color-coded according to their Item Type.

*Hint: Seaborn's lmplot is the easiest way to generate the scatter plot.*

```python
sns.lmplot(x="RetailSales", y="RetailTransfers", hue="ItemType", data=data, height=6, aspect=2);
#https://seaborn.pydata.org/generated/seaborn.lmplot.html
```

## 8. Create a scatter matrix using all the numeric fields in the data set with the plot points color-coded by Item Type.

*Hint: Seaborn's pairplot may be your best option here.*

```python
sns.pairplot(data, hue="ItemType")
#https://seaborn.pydata.org/generated/seaborn.pairplot.html
```
