```python
import matplotlib.pyplot as plt
import os
from string import ascii_letters
import numpy as np
import pandas as pd
import seaborn as sns
from datetime import datetime
import bar_chart_race as bcr
import re
import pymongo
from pandas.io.json import json_normalize
import folium
```

```python
cliente=pymongo.MongoClient()
```

```python
companies = cliente['companies']
companies.list_collection_names()
```

```python
colec = companies['companies']
colec
```

```python
keys = colec.find({})
keys[0].keys()
```

```python
colec_biotech = list(colec.find({'$and':[{'offices':{'$not':{'$size':0}}},
                                             {'category_code':'biotech'}]},
                                    {'_id':0,'name':1,'category_code':1,'offices':1}))
```

```python
print(len(colec_biotech))
```

```python
all_companies=pd.DataFrame(colec_biotech)
all_companies
```

```python
offices=[pd.json_normalize(i) for i in all_companies['offices']]
```

```python
offices
```

```python
offices[0]
```

```python
len(offices)
```

```python
for i, item in enumerate(offices):
    item['company'] = all_companies['name'][i]
```

```python
offices[0]
```

```python
biotechcos=[i[['company','country_code','state_code','city','address1','address2','zip_code','latitude','longitude']] for i in offices]
```

```python
biotechcos[0]
```

```python
biocomps=pd.DataFrame()
```

```python
for i in range(len(biotechcos)):
    biocomps = pd.concat([biocomps, biotechcos[i]])
```

```python
biocomps
```

```python
can=biocomps[biocomps['country_code']=='CAN']
can
```

```python
len(can)
```

```python
can['state_code'].unique()
```

```python
can['city'].unique()
```
