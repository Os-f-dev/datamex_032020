### Key Knowledge Points

1.-El lenguaje informático tiene vocabularios pequeños y sigue una estructura, el lenguaje natural (humano) tiene vocabularios diversos, significados diferentes, acentos, pronunciación incorrecta, se puede decir que hay errores en el lenguaje natural (humano).

2.-Porque no se le puede dar un diccionario de todas las frases que puedan existir a una computadora.

3.-Es que hay muchas palabras que tienen varios significados o que pueden ser verbo o sustantivo.

4.-Es una estructura que etiqueta las partes de un discurso y también nos da cómo se construye la oración.

5.-Los de la primera generación consistían en reglas hechas a mano y los modernos de PNL se basan en técnicas de aprendizaje automático.

6.-Codificando cientos de reglas que mapean lo que un usuario puede decir y cómo debe responder un programa.

7.-Reconocen Porque los datos se almacenan en una red de información semántica, donde las entidades están vinculadas entre sí.

8.-Es hacer que una computadora obtenga palabras del sonido, utilizando redes neuronales profundas para obtener una mejor precisión.

9.-Utiliza espectrogramas, procesados con el algoritmo de la transformada rápida de Fourier.

10.-Es generar lenguaje hablado por máquina sobre la base de la entrada escrita.

11.-Las interfaces de usuario de voz crean un ciclo de retroalimentación positiva al obtener más datos para entrenar los sistemas, lo que permite una mayor precisión.

### Key Knowledge Points 2

12.-Es derivar información significativa del texto en lenguaje natural.

13.-Es identificar y categorizar computacionalmente las opiniones expresadas en un texto, para determinar si es positiva, negativa o neutral.

14.-Resumen de texto, chatbots, análisis sentimental, clasificación automática de textos, generación de texto reconocimiento de voz, traducción automática.

15.-Tokenización, derivación, lematización, etiquetas POS, reconocimiento de entidades con nombre y fragmentación.

16.-Es romper cadenas en sus palabras individuales.

17.-Es normalizar las palabras en su forma básica.

18.-Es un proceso lingüístico que consiste en, dada una forma flexionada, hallar el lema correspondiente.

19.-La etiqueta gramatical de la palabra cómo nombre, verbo, etc.

20.-Es el proceso de detección de nombres cómo organizaciones, personas, etc.

21.-Recolecta piezas de información y las agrupa en piezas más grandes.
