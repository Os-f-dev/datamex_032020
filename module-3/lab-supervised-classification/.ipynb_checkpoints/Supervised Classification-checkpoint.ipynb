# Supervised classification

In the data.csv there are letters (uppercases and lowercases) and numbers, 28x28 pixels in a row format.

* First, you need to know which labels are which, meaning you need to visualize some data to realize which number labels represents a letter, or a number.

```python
# Import libraries
import zipfile
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter('ignore')

#dividir el dataframe
from sklearn.model_selection import train_test_split as tts

#modelos
from sklearn.linear_model import LogisticRegression as logreg
from sklearn.linear_model import SGDClassifier as SGDC
from sklearn.neighbors import KNeighborsClassifier as KNNC
from sklearn.ensemble import GradientBoostingClassifier as GBC
from lightgbm import LGBMClassifier as LGBMC
from xgboost import XGBClassifier as XGBC
from catboost import CatBoostClassifier as CATB
```

```python
# Your code here
zf = zipfile.ZipFile('data_all.zip')
df = pd.read_csv(zf.open('data_all.csv'))
```

```python
'''Muestra de la tabla (se ve que hay una columna "e" (etiqueta) y 784 columnas de pixeles (28x28),
cada fila es una imagen).'''
df.head()
```

```python
# Evaluamos qué etiquetas hay, se ven que son números del 0 al 61 (equivalente a 27 miyus, 27 mayus y 10 números)
df["e"].unique()
```

```python
#ploteamos una imagen de cada etiqueta "e" (se tiene que convertir el df a array)
plt.figure(figsize=(20,20))
for i in range(62):
    imagenes = np.array(df[df.e==i].drop("e", axis=1).T)
    plt.subplot(10,10,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(imagenes[:,1].reshape(28, 28).T, cmap=plt.cm.bone, interpolation='nearest')
    plt.xlabel(i)
plt.show()
```

```python
'''Insights:
-las etiquetas de números van del 0 al 9
-las etiquetas de mayusculas van del 10 al 35
-las etiquetas de minusculas van del 36 al 61'''
```

### Predecir

```python
# Generamos la lista de modelos a  probar
lista_m = [logreg]

# función para probar los modelos
def modeling_testing(lista_modelos, data):
    X_train, X_test, y_train, y_test = tts(data.loc[:, data.columns != "e"], data["e"])
    for i in lista_modelos:
        modelo = i()
        modelo.fit(X_train, y_train)
        train_score = modelo.score(X_train, y_train)
        test_score = modelo.score(X_test, y_test)
        print('Modelo:',str(i).split(".")[-1])
        print('Train score:', train_score,'\nTest score:',test_score,'\n')
```

* Now, try to train a classifier model to predict the uppercases. Use every single model you know for classification.

```python
mayus = df[(df["e"]>=10)&(df["e"]<=35)]

modeling_testing(lista_m,mayus)
```

* Try to do the same thing with lowercases.

```python
# Your code here
minus = df[(df["e"]>=36)]

modeling_testing(lista_m,minus)
```

* Try to do the same thing with numbers.

```python
# Your code here
num = df[(df["e"]<=9)]

modeling_testing(lista_m,num)
```
