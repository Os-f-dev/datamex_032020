Complete the exercises below to solidify your knowledge and understanding of unsupervised learning model evaluation.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

from sklearn import datasets

data = datasets.load_wine()

X = pd.DataFrame(data["data"], columns=data["feature_names"])
y = pd.Series(data["target"])
```

## 1. Train a KMeans clustering model on the data set using 8 clusters and compute the silhouette score for the model.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
kmeans = KMeans(n_clusters = 8)
kmeans.fit(X)

y1_pred= kmeans.predict(X)

silhouette_score(X, y1_pred)
```

## 2. Train a KMeans clustering model on the data set using 5 clusters and compute the silhouette score for the model.

```python
kmeans = KMeans(n_clusters = 5)
kmeans.fit(X)

y2_pred= kmeans.predict(X)

silhouette_score(X, y2_pred)
```

## 3. Train a KMeans clustering model on the data set using 3 clusters and compute the silhouette score for the model.

```python
kmeans = KMeans(n_clusters = 3)
kmeans.fit(X)

y3_pred= kmeans.predict(X)

silhouette_score(X, y3_pred)
```

## 4. Use elbow curve visualizations to see if you can determine the best number of clusters to use.

The Yellowbrick library has 3 metrics that you can plot using the `metric` parameter:

- **distortion**: mean sum of squared distances to centers
- **silhouette**: mean ratio of intra-cluster and nearest-cluster distance
- **calinski_harabaz**: ratio of within to between cluster dispersion

```python
from yellowbrick.cluster import KElbowVisualizer
```

```python
modelo =KMeans()

visual = KElbowVisualizer(modelo, k = (1,10), metric = 'distortion')

visual.fit(X)

visual.poof();
```

```python
modelo =KMeans()

visual = KElbowVisualizer(modelo,metric = 'silhouette', k = (2,12), )

visual.fit(X)

visual.poof();
```

```python
modelo =KMeans()

visual = KElbowVisualizer(modelo, metric = 'calinski_harabasz', k = (2,12))

visual.fit(X)

visual.poof();
```

## 5. Try performing the same elbow tests with an AgglomerativeClustering model and compare the results you get to the KMeans results.

```python
from sklearn.cluster import AgglomerativeClustering as AC
```

```python
modelo2 =AC()

visual2 = KElbowVisualizer(modelo2, k = (1,10), metric = 'distortion')

visual2.fit(X)

visual2.poof();
```

## 6. Create and plot a scatter matrix showing how the clusters are grouped across all the different combinations of variables in the data.

Use the model and number of clusters that returned the best result above.

```python
import seaborn as sns
```

```python

kmeans = KMeans(n_clusters = 3)
kmeans.fit(X)

y_pred= kmeans.predict(X)

matriz = pd.DataFrame(X, columns=X.columns)
matriz['pred']= y_pred

```

```python
sns.pairplot(matriz, hue="pred")
```

```python
pd.plotting.scatter_matrix(matriz)
```

## 7. Apply a PCA transform and plot the first two principle components with the plot point colors determined by cluster.

```python
from sklearn.decomposition import PCA
```

```python
modelopca= PCA(n_components = 2).fit_transform(X)
```

```python
dfpca =pd.DataFrame(modelopca, columns =['pca1','pca2'])
```

```python
dfpca['pred']= y_pred
```

```python
sns.scatterplot(dfpca['pca1'],dfpca['pca2'], hue = dfpca['pred'])
```

## 8. Generate a series of t-SNE plots showing the clusters at a variety of perplexities.

```python
from sklearn.manifold import TSNE
```

```python
X_embedded = TSNE(n_components=2).fit_transform(X)

sns.scatterplot(X_embedded[:,0],X_embedded[:,1], hue = dfpca['pred'])
```

```python
fig,graficas = plt.subplots(figsize = (15,6), nrows = 1, ncols = 3)

plt.sca(graficas[0])
X0_embedded = TSNE(n_components=2, perplexity=30).fit_transform(X)
sns.scatterplot(X0_embedded[:,0],X0_embedded[:,1], hue = dfpca['pred'])
plt.title(['perplexity = 30'])

plt.sca(graficas[1])
X1_embedded = TSNE(n_components=2, perplexity=60).fit_transform(X)
sns.scatterplot(X1_embedded[:,0],X1_embedded[:,1], hue = dfpca['pred'])
plt.title(['perplexity = 60'])

plt.sca(graficas[2])
X2_embedded = TSNE(n_components=2, perplexity=90).fit_transform(X)
sns.scatterplot(X2_embedded[:,0],X2_embedded[:,1], hue = dfpca['pred'])
plt.title(['perplexity = 90'])

plt.show();
```
