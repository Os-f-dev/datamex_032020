```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from sklearn import cluster
```

```python
from sklearn import datasets

data = datasets.load_wine()

X = pd.DataFrame(data["data"], columns=data["feature_names"])
y = pd.Series(data["target"])

X.head()
```

## 1. Train a KMeans clustering model on the data set using 8 clusters and compute the silhouette score for the model.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

kmeans = KMeans(n_clusters=8)
km_clus = kmeans.fit(X)
y1_pred= km_clus.predict(X)
silhouette_score(X, y1_pred)
```

```python
fig, axs =plt.subplots(13,13, figsize=(13,13))
for i in range(13):
    for j in range(13):
        axs[i,j].scatter(X.iloc[:,i],X.iloc[:,j],c=y1_pred)
```

```python
plt.scatter(X.iloc[:,12],X.iloc[:,2],c=y1_pred)
```

## 2. Train a KMeans clustering model on the data set using 5 clusters and compute the silhouette score for the model.

```python
kmeans5 = KMeans(n_clusters=5)
km_clus5 = kmeans.fit(X)
y5_pred= km_clus5.predict(X)
silhouette_score(X, y5_pred)
```

```python
plt.scatter(X.iloc[:,12],X.iloc[:,4],c=y5_pred)
```

## 3. Train a KMeans clustering model on the data set using 3 clusters and compute the silhouette score for the model.

```python
kmeans3 = KMeans(n_clusters=3)
km_clus3 = kmeans3.fit(X)
y3_pred= km_clus3.predict(X)
silhouette_score(X, y3_pred)
```

```python
y3_pred
```

```python
plt.subplots(figsize=(14,8))
colors = np.array(['black', 'silver', 'red'])

legend_e = [Line2D([0],[0], color = c, marker ='o', linestyle = '', label = f'Clase{i}') for i,c in enumerate(colors)]

plt.scatter(X.iloc[:,0], X.iloc[:,12], color= colors[y3_pred])
plt.legend(handles = legend_e, loc = 'upper right')
```

## 4. Use elbow curve visualizations to see if you can determine the best number of clusters to use.

The Yellowbrick library has 3 metrics that you can plot using the `metric` parameter:

- **distortion**: mean sum of squared distances to centers
- **silhouette**: mean ratio of intra-cluster and nearest-cluster distance
- **calinski_harabaz**: ratio of within to between cluster dispersion

```python
from yellowbrick.cluster import KElbowVisualizer
```

```python
modelo = KMeans()

visual = KElbowVisualizer(modelo, k = (1,10))

visual.fit(X)

visual.poof();
```

```python
# Pareciera que la curva ms pronunciada es en el punto 2, pero en el modelo nos dice que lo mejor es 3
```

## 5. Try performing the same elbow tests with an AgglomerativeClustering model and compare the results you get to the KMeans results.

```python
modelo = cluster.AgglomerativeClustering()

visual = KElbowVisualizer(modelo, k = (1,10))

visual.fit(X)

visual.poof();
```

```python
single = cluster.AgglomerativeClustering(n_clusters=3)
```

```python
single.fit(X)
yagglo1_pred = single.labels_
yagglo1_pred
```

```python
plt.subplots(figsize=(14,8))
colors = np.array(['red', 'green','silver'])

legend_e = [Line2D([0],[0], color = c, marker ='o', linestyle = '', label = f'Clase{i}') for i,c in enumerate(colors)]

plt.scatter(X.iloc[:,0], X.iloc[:,12], color= colors[yagglo1_pred])
plt.legend(handles = legend_e, loc = 'upper right')
```

## 6. Create and plot a scatter matrix showing how the clusters are grouped across all the different combinations of variables in the data.

Use the model and number of clusters that returned the best result above.

```python
import seaborn as sns
X['cluster']=yagglo1_pred
sns.pairplot(X,hue="cluster",diag_kind='hist')
```

## 7. Apply a PCA transform and plot the first two principle components with the plot point colors determined by cluster.

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)

principales = pca.fit_transform(X)
prici2=pd.DataFrame(principales,columns=['Component 1','Component 2'])
prici2.head()
```

```python
sns.scatterplot(data=prici2,x="Component 1", y="Component 2", hue=yagglo1_pred) 
```
